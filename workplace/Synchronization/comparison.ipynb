```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
img1 = np.random.randn(100, 100)
img2 = np.random.randn(100, 100)
imgs = [img1, img2]
```

```python
angle_range = np.linspace(-2, 2, 100)
fig, ax = plt.subplots()
im1 = ax.imshow(imgs[0])
def update(frame):
    # 
    im1.set_data(imgs[frame%2])
    return [im1]

# 
ani = FuncAnimation(fig, update, frames=len(angle_range), interval=50, blit=True)

# 
plt.show()
```
